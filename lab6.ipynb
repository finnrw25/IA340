```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# assuming the dataframe is named as df
df['income'] = df['income'].astype(float)  # making sure income is a float
df = df.sort_values(['name', 'year'])  # sorting by state and year

# calculating the year over year growth rate
df['income_growth'] = df.groupby('name')['income'].pct_change()

# plotting line chart
plt.figure(figsize=(10, 6))
for state in df['name'].unique():
    subset = df[df['name'] == state]
    plt.plot(subset['year'], subset['income_growth'], label=state)
plt.xlabel('Year')
plt.ylabel('Income growth rate')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# assuming the dataframe is named as df
df['income'] = df['income'].astype(float)  # making sure income is a float
df = df.sort_values(['name', 'year'])  # sorting by state and year

# calculating the year over year growth rate
df['income_growth'] = df.groupby('name')['income'].pct_change()

# creating a pivot table to plot
pivot_df = df.pivot(index='year', columns='name', values='income_growth')

# plotting bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('Year')
plt.ylabel('Income growth rate')
plt.show()
```
